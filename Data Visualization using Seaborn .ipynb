### Bar plot

- Use size contrast to compare two or more variables.
- Bar charts, seen often in the popular press, are a common visual tool for displaying a single categorical variable. 

```python
sns.barplot(x="species",y="body_mass_g",data=penguins)
```

```python
sns.catplot(x="species",y="body_mass_g",data=penguins, kind="bar")
```

##### as we can see the small black line in the middle of the bar -----> this is mean and we can remove it by ci = None, ci refered to condidence intervals 
- the line of ci show 95% confidence intervals for the mean 


```python
# removing the condidence intervals, we can also chose what it represents  
sns.catplot(x="species",y="body_mass_g",data=penguins, kind="bar",ci= None)
```

#### we can use estimator to change the black line to show any measure 

```python
# the black line now show the std of each bar
sns.catplot(x="species",y="body_mass_g",data=penguins, kind="bar",estimator= np.std)
```

```python
# change the estimator to shows the max
sns.catplot(x="species",y="body_mass_g",data=penguins, kind="bar",estimator= np.max)
```

```python
# change the estimator to shows the min
sns.catplot(x="species",y="body_mass_g",data=penguins, kind="bar",estimator= np.min)
```

```python
#we can check of the valuse 
penguins.groupby('species').body_mass_g.describe()
```

#### we can make the bar Horizontal by swinping the x and Y values

```python
sns.catplot(x="body_mass_g",y="species",data=penguins, kind="bar",estimator= np.min)
```

### hue, hue_order and order

```python
#adding hue
sns.catplot(x="species",y="body_mass_g",data=penguins, kind="bar",ci= None,hue= "sex")
```

```python
# order the bars from high to low and change the hue_order
sns.catplot(x="species",y="body_mass_g",
            data=penguins, 
            kind="bar",
            ci= None,
            hue= "sex",
            hue_order=["Female","Male"],
            order=["Gentoo","Adelie","Chinstrap"])
```

### Count plot

- A bar plot presents categorical data with rectangular bars with lengths proportional to the values that they represent.
- A bar plot shows comparisons among discrete categories .
- One axis of the plot shows the specific categories being compared, and the other axis represents a measured value.

```python
penguins.dtypes
```

```python
sns.catplot(x="island",
           data=penguins ,
           kind="count",
           hue="sex")
```

#### we can order the values by making a list that contains the order and then pass it to order 

```python
 penguins.island.value_counts().index
```

### Box plot

- Shows the distribution of quantitative data
- Seemedian, spread,skewness,andoutliers
- Facilitates comparisons between groups



```python
#loading data to practice 
cars= sns.load_dataset("mpg")
cars.head()
```

```python
cars.shape
```

```python
#cleaning data from nan vaules
cars=cars.dropna()
cars.shape
#we removed 6 nan vaules
```

```python
#we will deal with cylinders column so we need to know the values of him
cars.cylinders.value_counts()
```

```python
#as the odd number of cylinders is too low so we will agnore values
cars=cars[cars.cylinders.isin([4,6,8])]
cars.cylinders.value_counts()
```

```python
# lets see how the cars consumption per galion (mpg)in each country (origin)
sns.catplot(x="origin",
       y="mpg",
       data=cars,
        kind="box")
```

<b> as we can see outliers we have good ways to deal with them 
   

```python
# first is to remove them by "sym"= ""
sns.catplot(x="origin",
       y="mpg",
       data=cars,
        kind="box",
           sym="")
```

```python
# boxplot likes other Categorical plots in hue and order 
sns.catplot(x="origin",
       y="mpg",
       data=cars,
        kind="box",
        hue="cylinders",
        order=["europe","japan","usa"])
```

#### second way to deal with outliers is by whiskers -----> whis

- By default, the whiskers extend to 1.5* the interquartile range
- we can change them to include outliers or include all vaules

```python
# we change the whis to include the all data 
sns.catplot(x="origin",
       y="mpg",
       data=cars,
        kind="box",
        whis=[0,100])
```

```python
# here we make them extand to 2.0 instead of 1.5 * IQR 
sns.catplot(x="origin",
       y="mpg",
       data=cars,
        kind="box",
        whis=2.0)
```

### point plot

###  what is the differnce between  point plot and bar plot

<b> both show :
    
- Points show mean of quantitative variable
- Vertical lines show 95% condence intervals

```python
sns.catplot(x="species",y="body_mass_g",
            data=penguins, 
            kind="point",
            hue= "sex",
            hue_order=["Female","Male"],
            order=["Gentoo","Adelie","Chinstrap"])
```

```python
sns.catplot(x="species",y="body_mass_g",
            data=penguins, 
            kind="bar",
            hue= "sex",
            hue_order=["Female","Male"],
            order=["Gentoo","Adelie","Chinstrap"])
```

```python
# if we want to remove the lines conneted to the points ------> join = false
sns.catplot(x="species",y="body_mass_g",
            data=penguins, 
            kind="point",
            hue= "sex",
            hue_order=["Female","Male"],
            join = False,
            order=["Gentoo","Adelie","Chinstrap"])
```

```python
# also we can use estimator like bar plot
sns.catplot(x="species",y="body_mass_g",
            data=penguins, 
            kind="point",
            hue= "sex",
            hue_order=["Female","Male"],
            estimator= np.max,
            order=["Gentoo","Adelie","Chinstrap"])
```

```python
# in box plot we has whis here we have capsize 
sns.catplot(x="species",y="body_mass_g",
            data=penguins, 
            kind="point",
            hue= "sex",
            hue_order=["Female","Male"],
            capsize= 0.5,
            order=["Gentoo","Adelie","Chinstrap"])
```

```python
# we also can remove ci with set ci = None
sns.catplot(x="species",y="body_mass_g",
            data=penguins, 
            kind="point",
            hue= "sex",
            hue_order=["Female","Male"],
            ci=None,
            order=["Gentoo","Adelie","Chinstrap"])
```

# Quantitative plots

### histogram plot

- A histogram is the graphical representation of data where data is grouped into continuous number ranges and each range corresponds to a vertical bar


```python
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species")
```

<b> in histogram we have bins, binwidth and binrange so what are these and what are the difference
    
- first, bins -------> it means the number of bars that are display so if i want to display 20 bar i will set bins = 20
- second, binwdth refers to the scale of each bin, if i want each bin scale 10 i will set binwidth = 10
- third, binrange it means the range that will display in histogram , if range is between 100 -200 and the data is between 10- 70 the histogram will not appear beacause he is out of the range

```python
#bins =30
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            bins=30)
```

```python
# bin width = 10 
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            binwidth=10)
```

```python
# the histogram doesn't appear because he is out of range
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            binrange=(70,150))
```

<b> we also can change the measure of histogram to be any measure as we like. using ------> stat 

```python
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            stat= "density")
```

<b> we can change the view of the histogram by using------>  muliple  it has options like {layer,dodge,stack,fill}


```python
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            multiple="layer")
```

```python
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            multiple="dodge")
```

```python
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            multiple="fill")
```

```python
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            multiple="stack")
```

#### Changing the figure style

- sns.set_style()
- it have five options [white,dark,whitegrid,darkgrid,ticks]

```python
sns.set_style("darkgrid")
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            stat= "density")
```

```python
sns.set_style("whitegrid")
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            stat= "density")
```

```python
sns.set_style("dark")
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            stat= "density")
```

```python
sns.set_style("ticks")
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            stat= "density")
```

#### Changing the figure style

- Figure "palette" changes the color of the main elements of the plot
- <b> sns.set_palette() 
- Use preset palettes or create a custom palette
    
    


    
    



```python
sns.set_palette("RdBu")
sns.histplot(x="bill_length_mm",
            data=penguins,
              hue= "species",
            bins=30)
```

#### Changing the scale

- Figure "context" changes the scale of the plot elements andl abels
- sns.set_context()
- Smallest to largest :"paper","notebook","talk","poster"


```python
sns.set_style("whitegrid")
sns.set_palette("Greys")
sns.set_context("poster")
sns.histplot(x="bill_length_mm",
            data=penguins,
            hue= "species",
            stat= "density")
```
